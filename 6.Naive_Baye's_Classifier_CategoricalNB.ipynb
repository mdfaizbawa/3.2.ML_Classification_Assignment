```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
#howmany class in Target_Value. Balanced or imbalanced data
dataset['classification_yes'].value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e','rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes','dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
independent.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=1/3,random_state=0)
```

```python
from sklearn.naive_bayes import CategoricalNB
classifier =CategoricalNB()
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,y_pred)
print(cm)
print(clf_report)
```
