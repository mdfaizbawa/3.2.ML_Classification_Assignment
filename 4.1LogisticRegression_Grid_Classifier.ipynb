```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
#howmany class in Target_Value. Balanced or imbalanced data
dataset['classification_yes'].value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e','rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes','dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
independent.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=1/3,random_state=0)
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
X_train=sd.fit_transform(X_train)
X_test=sd.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
grid_param={'penalty':['l2'],'solver':['lbfgs','liblinear','newton-cg','newton-cholesky','saga','sag']}
grid=GridSearchCV(LogisticRegression(),grid_param, refit=True, verbose=3, n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
print("Confusion Matrix value for best parameter:{}".format(grid.best_params_))
```

```python
res=grid.cv_results_
table=pd.DataFrame.from_dict(res)
table
```

```python
y_pred=grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,y_pred)
print(cm)
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc=roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
roc
```

```python
import pickle
filename="finalized_model_LogisticRegressoin_classifier.sav"
pickle.dump(grid,open(filename,'wb'))
```
